```python
import cv2
```

```python
import os
from keras.preprocessing import image
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
categories = ['with_mask','without_mask']
datadir = "D:\Data Science Project\Face Mask Detection\data"
```

```python
data=[]
for category in categories:
    path = os.path.join(datadir,category)
#     print(path)
    
    label = categories.index(category)
#     print(label)
    
    
    for file in os.listdir(path):
#         print(file)
# # #         break
        img_path = os.path.join(path,file)
#         print(img_path)
        img =cv2.imread(img_path)
#         print(img)
        img =cv2.resize(img,(224,224))
#         print(img)
#         img = img[0]
#         print(img.shape)
        data.append([img,label])

# print(data)
```

```python
import random
random.shuffle(data)
```

### Split into X & Y

```python
X = []
Y = []
for features,label in data:
    X.append(features) #2d
    Y.append(label)  # 1d
```

```python
print(len(X))
print(len(Y))
```

```python
x = np.array(X)
y = np.array(Y)
# print(x)
```

```python
print(y)
print(x.shape)
print(y.shape)
```

```python
# x scaling
x=x/255
```

### Splite into train and test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
```

### Import VGG16

```python
from keras.applications import VGG16
vgg=VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential
from keras import layers
model=Sequential()
```

### Remove prediction layer

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.summary()
```

### Adding last layer

```python
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
cap = cv2.VideoCapture(0)
```

### Prediction

```python
def detect_face_mask(img):
    
    y_pred =  np.round(model.predict(img.reshape(1,224,224,3)))
    
    if y_pred[0][0] == 0:
        print("With Mask")
    else:
        print("Without Mask")
    
#     return y_pred[0][0]
```
